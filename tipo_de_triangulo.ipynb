#**Tipo de triángulo: acutángulo, obtusángulo o rectángulo**
Conociendo la longitud sus lados, podemos averiguar qué tipo de triángulo tenemos:


1.   $ladoMenor^2 + ladoMediano^2 = ladoMayor^2$ $->$ **Es un triángulo rectángulo**
2.   $ladoMenor^2 + ladoMediano^2 < ladoMayor^2$ $->$ **Es un triángulo obtusángulo**
1.   $ladoMenor^2 + ladoMediano^2 > ladoMayor^2$ $->$ **Es un triángulo acutángulo**

Usando sentencias if_elif_else debemos determinar qué tipo de triángulo tenemos

```python
#Introducimos los valores de los lados

ladoMenor = int(input("Introduce la longitud del lado menor: "))
ladoMediano = int(input("Introduce la longitud del lado menor: "))
ladoMayor = int(input("Introduce la longitud del lado menor: "))

#implementamos las declaraciones if_elif_else donde se imprima el tipo de triángulo

if ladoMenor**2+ladoMediano**2==ladoMayor**2:
  print("Es un triángulo rectángulo")

elif ladoMenor**2+ladoMediano**2<ladoMayor**2:
  print("Es un triángulo obtusángulo")

else:
  print("Es un triángulo acutángulo")

```
